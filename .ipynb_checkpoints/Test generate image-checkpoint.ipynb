```python
import cv2
import os
import numpy as np
```

```python
dataset_directory = "raw_data/EpicureAi.v3-resize-512x512-reoriented.yolov8/train/images"
path_new_images = "raw_data/EpicureAi.v3-resize-512x512-reoriented.yolov8/train"
```

```python
def rotate_image(image, angle):
    height, width = image.shape[:2]
    rotation_matrix = cv2.getRotationMatrix2D((width/2, height/2), angle, 1)
    rotated_image = cv2.warpAffine(image, rotation_matrix, (width, height))
    return rotated_image
```

```python
def flip_image(image, flipCode):
    return cv2.flip(image, flipCode)
```

```python
def adjust_brightness(image, value):
    hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
    h, s, v = cv2.split(hsv)
    v = cv2.add(v, value)
    v[v > 255] = 255
    v[v < 0] = 0
    final_hsv = cv2.merge((h, s, v))
    return cv2.cvtColor(final_hsv, cv2.COLOR_HSV2BGR)
```

```python
new_images_count = 0
target_images_count = 3000
transformations = [rotate_image, flip_image]  # Vous pouvez ajouter plus de fonctions ici

for filename in os.listdir(dataset_directory):
    if new_images_count >= target_images_count:
        break
    if filename.endswith(".jpg") or filename.endswith(".png"):
        file_path = os.path.join(dataset_directory, filename)
        image = cv2.imread(file_path)
        for transform in transformations:
            for angle in range(0, 360, 15):  # Exemple de rotation à différents angles
                new_image = transform(image, angle)
                new_file_path = os.path.join(path_new_images, f"new_image_{new_images_count}.jpg")
                cv2.imwrite(new_file_path, new_image)
                new_images_count += 1
                if new_im ages_count >= target_images_count:
                    break
```
